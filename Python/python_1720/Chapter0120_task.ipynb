__[과제 1: 초기화 메서드 __init__]__

- stock
    - "팥붕어빵": 10
    - "슈크림붕어빵": 8
    - "초코붕어빵": 5

- prices
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

```python
def __init__(self):
    self.stock = {"팥붕어빵": 10,
                  "슈크림붕어빵" : 8,
                  "초코붕어빵": 5
    }
    self.prices = {"팥붕어빵: 1000",
                   "슈크림붕어빵: 1200",
                   "초코붕어빵: 1500"
    }
    self.sales = {"팥붕어빵: 0",
                  "슈크림붕어빵: 0",
                  "초코붕어빵: 0"
    }
```

__[과제 2: 재고 확인 메서드 check_stock]__

목표: check_stock 함수를 구현하여 현재 붕어빵 재고를 출력합니다.

```python
def check_stock(self):
    print("현재 붕어빵 재고:")
    for bread, quantity in self.stock.items():
        print(f'{bread}: {quantity}개')
```

설명: bread_type과 bread_count를 매개변수로 받아, 재고를 확인하고 충분한 경우 재고를 감소시키고 판매량을 증가시킵니다. 재고가 부족하거나 잘못된 붕어빵 종류인 경우 적절한 메시지를 출력하세요.

```python
def procss_order(self, bread_type, bread_count):
    if self.stock[bread_type] >= bread_count:
        self.stock[bread_type] -= bread_count
        self.sales[bread_type] += bread_count
        print(f'{bread_type} {bread_count}개 주문이 완료되었습니다.')
        if bread_type not in bread_count <= 0:
            print("잘못된 주문입니다")
```

__[과제 4: 관리자 모드 메서드 admin_mode]__

목표: admin_mode 메서드를 구현하여 관리자 모드에서 재고를 추가할 수 있도록 합니다.

설명: bread_type과 additional_stock을 입력받아 재고를 추가합니다. 잘못된 입력 값이 들어온 경우 적절한 메시지를 출력하세요.

```python
def admin_mode(self):
    while True:
        bread_type = input("추가할 붕어빵 종류를 입력하세요 (팥붕어빵, 슈크림붕어빵, 초코붕어빵) 종료를 원하시면 '종료'를 입력하세요:")
        if bread_type == "종료":
            break
        try:
            additional_stock = int(input(f"{bread_type}추가할 재고량을 입력하세요: "))
            self.stock[bread_type] += additional_stock
            print(f'{bread_type}의 재고가 {additional_stock}개 추가되었습니다.')
            self.check_stock()
        except ValueError:
                print("잘못된 입력입니다. 숫자를 입력해주세요.")
        except KeyError:
                print("메뉴에 없는 붕어빵 종류입니다.")
```


__[과제 5: 매출 계산 메서드 calculate_total_sales]__

목표: calculate_total_sales 메서드를 구현하여 총 매출을 계산하고 출력합니다.

설명: self.sales와 self.prices를 사용하여 총 매출을 계산하고 출력하세요.

```python
def calculate_total_sales(self):
    total_sales = sum(self.sales[bread] * self.prices[bread]
                      for bread in self.sales)
    print(f'총 매출: {total_sales}원')
    print('거래 내역')
    for bread in self.sales:
        print(f'{bread} 판매: {self.sales[bread]}개, 매출: {self.sales[bread] * self.prices[bread]}원')
```

__[과제 7: 최종 BungeoppangShop 클래스]__

함수로 리팩토링한 "붕어빵 자동 판매 시스템" 작성 후 실행합니다.

```python
class BungeoppangShop:
    def __init__(self):
        self.stock = {"팥붕어빵": 10,
                      "슈크림붕어빵" : 8,
                      "초코붕어빵": 5
        }
        self.prices = {"팥붕어빵": 1000,
                       "슈크림붕어빵": 1200,
                       "초코붕어빵": 1500
        }
        self.sales = {"팥붕어빵": 0,
                      "슈크림붕어빵": 0,
                      "초코붕어빵": 0
        }

    def check_stock(self):
        print("현재 붕어빵 재고:")
        for bread, quantity in self.stock.items():
            print(f'{bread}: {quantity}개')

    def process_order(self, bread_type, bread_count):
        try:
            if self.stock[bread_type] >= bread_count:
                self.stock[bread_type] -= bread_count
                self.sales[bread_type] += bread_count  # 판매량 업데이트
                print(f"{bread_type} {bread_count}개를 판매했습니다.")
            else:
                print(f"죄송합니다. {bread_type} 재고가 부족합니다.")
        except KeyError:
            print("존재하지 않는 붕어빵 종류입니다.")

    def admin_mode(self):
        while True:
            bread_type = input("추가할 붕어빵 종류를 입력하세요 (팥붕어빵, 슈크림붕어빵, 초코붕어빵) 종료를 원하시면 '종료'를 입력하세요:")
            if bread_type == "종료":
                break
            try:
                additional_stock = int(input(f"{bread_type}추가할 재고량을 입력하세요: "))
                self.stock[bread_type] += additional_stock
                print(f'{bread_type}의 재고가 {additional_stock}개 추가되었습니다.')
                self.check_stock()
            except ValueError:
                print("잘못된 입력입니다. 숫자를 입력해주세요.")
            except KeyError:
                print("메뉴에 없는 붕어빵 종류입니다.")

    def calculate_total_sales(self):
        total_sales = sum(self.sales[bread] * self.prices[bread] for bread in self.sales)
        print(f'총 매출: {total_sales}원')
        print('거래 내역')
        for bread in self.sales:
            print(f'{bread} 판매: {self.sales[bread]}개, 매출: {self.sales[bread] * self.prices[bread]}원')

def main():
    shop_bp = BungeoppangShop()

    while True:
        mode_choose = input("모드를 선택해주세요 (주문모드,관리자모드) 종료를 원하시면 '종료'를 입력하세요:")

        if mode_choose == '종료':
            break

        if mode_choose == '관리자모드':
            shop_bp.admin_mode()
            continue

        if mode_choose == '주문모드':
            while True:
                bread_type = input("주문할 붕어빵 종류를 입력하세요 (팥붕어빵, 슈크림붕어빵, 초코붕어빵) 종료를 원하시면 '종료'를 입력하세요:")
                if bread_type == '종료':
                    break
                try:
                    bread_count = int(input("주문할 개수를 입력하세요: "))
                    shop_bp.process_order(bread_type, bread_count)
                except ValueError:
                    print('잘못된 입력입니다 숫자로 입력해주세요.')

                shop_bp.check_stock()

    shop_bp.calculate_total_sales()

if __name__ == "__main__":
    main()
```
