이는 코드베이스 전체에서 일관성 있는 스타일을 유지하고 코드를 더 쉽게 유지 보수할 수 있게 한다.

"블록 구분" 들여쓰기를 사용하면 코드 블록의 시작과 끝이 명시적으로 표시되므로 중괄호를 사용하는 언어에서

발생하는 오류 중 일부를 방지할 수 있다.

"코드 블록 강제" 들여쓰기는 코드 블록을 강제하므로 잘못된 코드 구조를 방지한다.

코드 블록이 끝나는 위치를 표현하기 위해 중괄호를 사용하는 것보다 들여쓰기를 사용하는 것이 간단하고

실수를 줄일 수 있다.

<span style="font-size:80%">[문제5]<br><br></span>
```print( )문의 기능 및 특징에 대해 설명해주세요```

<span style="color:red">답변</span>

print() 함수는 파이썬 내장함수 중 하나로 특별히 다른 모듈을 활용하지 않아도 사용할 수 있는 함수이며,

python에서 화면에 텍스트를 출력하는 기능을 수행하는 내장 함수이다.

기본적으로 문자열을 출력하는 데 사용되지만 다양한 타입의 데이터를 출력할 수 있다.

화면 출력:

print() 함수는 괄호 안의 값을 화면에 출력한다.

이 값을 여러 개 지정하면 각 값은 공백으로 구분되어 출력

서식화된 출력:

%나 format() 메서드를 사용하여 서식화된 출력을 지원

파일 출력:

print() 함수는 파일에도 출력할 수 있다.

file 매개변수를 사용하여 출력 대상 파일 객체를 지정

문자열 연결:

print() 함수를 사용하여 여러 값을 하나의 문자열로 연결하여 출력

자동 줄 바꿈:

기본적으로 print() 함수는 각 호출 사이에 자동으로 줄 바꿈을 추가

이 동작을 변경하려면 end 매개변수를 사용할 수 있다.

디버깅 및 테스트:

print() 함수는 간단한 디버깅 및 테스트에 유용하게 사용된다.

변수의 값을 확인하거나 특정 코드 부분에서 실행되었는지 확인할 때 활용

sep 매개변수:

sep 매개변수를 사용하여 출력되는 값 사이에 구분자를 지정할 수 있다.
